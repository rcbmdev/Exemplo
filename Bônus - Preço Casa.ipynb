# Divisão Dados

```python
X_casas = base_casa.iloc[:, 5:6].values
X_casas
```

```python
y_casas = base_casa.iloc[:, 2].values
y_casas
```

```python
from sklearn.model_selection import train_test_split
X_casas_treino, X_casas_teste, y_casas_treino, y_casas_teste = train_test_split(X_casas, y_casas, test_size=0.3)
```

```python
X_casas_treino.shape, y_casas_treino.shape
```

```python
X_casas_teste.shape, y_casas_teste.shape
```

# Treinamento Modelo

```python
from sklearn.linear_model import LinearRegression
reg_casa = LinearRegression()
reg_casa.fit(X_casas_treino, y_casas_treino)
```

```python
reg_casa.intercept_
```

```python
reg_casa.coef_
```

```python
reg_casa.score(X_casas_teste, y_casas_teste)
```

# Prevendo valores

```python
reg_casa.predict([[1000]])
```

```python
previsoes = reg_casa.predict(X_casas_teste)
```

```python
previsoes
```

```python
grafico = px.scatter(x = X_casas_teste.ravel(), y = previsoes)
grafico.show()
```

```python
grafico1 = px.scatter(x = X_casas_treino.ravel(), y = y_casas_treino)
grafico2 = px.line(x = X_casas_teste.ravel(), y = previsoes)
grafico2.data[0].line.color = 'red'
grafico3 = go.Figure(data=grafico1.data + grafico2.data)
grafico3
```
